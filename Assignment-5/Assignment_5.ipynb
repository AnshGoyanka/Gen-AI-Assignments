```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A futuristic city at sunset, cyberpunk style, ultra detailed, digital art"

image = pipe(prompt).images[0]

image
```

```python
image.save("generated_art.png")
```

```python
prompts = [
    "Fantasy dragon flying over mountains, cinematic lighting",
    "Indian temple in space, sci fi art, highly detailed",
    "AI robot painting a sunset, digital illustration",
    "Underwater city with glowing plants, fantasy art"
]

for p in prompts:
    img = pipe(p).images[0]
    display(img)
```
