```python
from transformers import pipeline

generator = pipeline(
    task="text-generation",
    model="google/flan-t5-base"
)
```

```python
def ask_model(prompt):
    output = generator(prompt, max_length=200)
    return output[0]['generated_text']
```

```python
zero_shot_prompt = """
Classify sentiment as Positive, Negative or Neutral:

Sentence: I love this AI course.
"""

print("Zero Shot Result:")
print(ask_model(zero_shot_prompt))
```

```python
few_shot_prompt = """
Classify sentiment:

Sentence: I hate traffic.
Sentiment: Negative

Sentence: This movie is amazing.
Sentiment: Positive

Sentence: The food was average.
Sentiment: Neutral

Sentence: I really like this laptop.
Sentiment:
"""

print("Few Shot Result:")
print(ask_model(few_shot_prompt))
```

```python
interview_prompt = """
You are a career counselor.

Ask 3 questions to understand student's interest, skills and goals.
Then suggest 2 career options.
"""

print("Interview Approach:")
print(ask_model(interview_prompt))
```

```python
cot_prompt = """
Solve step by step:

A pen costs 20 rupees.
Shop gives 10% discount.
What is final price?
"""

print("CoT Result:")
print(ask_model(cot_prompt))
```

```python
tot_prompt = """
Problem: Increase sales of a new food delivery app.

Step 1: Suggest 3 strategies.
Step 2: Give pros and cons of each.
Step 3: Select best strategy and explain why.
"""

print("ToT Result:")
print(ask_model(tot_prompt))
```

```python
print("""
Comparison:

Zero Shot:
- No examples
- Fast
- Less context understanding

Few Shot:
- Uses examples
- Better accuracy

Interview Approach:
- Personalized outputs

CoT:
- Good reasoning clarity

ToT:
- Best for complex decision making
""")
```
